```python
import pandas as pd

# Example in Python (PyTorch/NumPy) side:
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

gsva_matrix = pd.read_csv("dataset/ccle_gsva_scores.csv", index_col=0)
gsva_matrix
```

```python
X_cell = scaler.fit_transform(gsva_matrix)
```

```python
X_cell
```

```python
X_cell = pd.DataFrame(X_cell, index=gsva_matrix.index, columns=gsva_matrix.columns)
```

```python
X_cell.to_csv("dataset/ccle_gsva_scores_normalized.csv")
```
