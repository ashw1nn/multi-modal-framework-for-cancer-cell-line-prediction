```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset/CCLE_RNAseq_rsem_genes_tpm_20180929.txt", sep="\t", index_col=0)
```

```python
df
```

```python
memory_MB = df.memory_usage(deep=True).sum() / (1024 ** 2)
memory_MB
```

```python
df.describe()
```

```python
df = df.drop(columns=["transcript_ids"], errors="ignore")
```

```python
# Remove genes with very low expression across all samples
df = df[df.mean(axis=1) > 1]
df
```

```python
df_log = np.log2(df + 1)
```

```python
df_log.describe()
```

```python
df_log.to_csv("dataset/ccle_expression_log.csv")
```

```python
df_t = df_log.T  # Now rows = cell lines, columns = genes
df_t
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
reduced = pca.fit_transform(df_t)

plt.scatter(reduced[:, 0], reduced[:, 1])
plt.title("PCA of CCLE RNA-seq")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
